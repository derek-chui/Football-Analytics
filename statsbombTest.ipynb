```python
MATCH_ID = 3942819
match_events_df = sb.events(match_id=MATCH_ID)
match_360_df = pd.read_json(f'/Users/derekchui/Documents/Programming/Personal/Projects/Analytics/open-data/data/three-sixty/{MATCH_ID}.json')
df = pd.merge(left=match_events_df, right=match_360_df, left_on='id', right_on='event_uuid', how='left')
```

```python
MAINOO = 99174
df = df[(df['player_id'] == MAINOO) & (df['type'] == 'Pass')].reset_index(drop=True)
df.location
```

```python
df[['x_start', 'y_start']] = pd.DataFrame(df.location.tolist(), index=df.index)
df[['x_end', 'y_end']] = pd.DataFrame(df.pass_end_location.tolist(), index=df.index)
```

```python
p = Pitch(pitch_type='statsbomb')
fig, ax = p.draw(figsize=(12,8))
df = df[0:1]
p.scatter(x=df['x_start'], y=df['y_start'], ax=ax)
p.lines(xstart=df['x_start'], ystart=df['y_start'], xend=df['x_end'], yend=df['y_end'], ax=ax, comet=True)
for x in df.iloc[0]['freeze_frame']:
    if x['teammate']:
        color = 'blue'
    else:
        color = 'red'
    p.scatter(x=x['location'][0], y=x['location'][1], ax=ax, c=color, alpha=1)
```
